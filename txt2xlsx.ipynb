```python
wb = openpyxl.load_workbook('freport3.xlsx')
sheet = wb.active
```

## initializing header for xlsx file


```python
sheet['A1'] ='ProcessId'
sheet['B1'] ='Image'
sheet['C1'] ='User'
sheet['D1'] ='UtcTime'
```

## Search for headers values and insert each value in its column

```python
flag = 2
for i, line in enumerate(open('freport3.txt')):
    for match in re.finditer(pattern, line):
        #print(flag)
        if match.group(1) == 'ProcessId':
            sheet['A'+str(flag)] = match.group(2)
        if match.group(1) == 'Image':
            sheet['B'+str(flag)] = match.group(2)
        if match.group(1) == 'User':
            val = match.group(2)
            t = val.split('"')[0].strip()
            print(t)
            sheet['C'+str(flag)] = t
        if match.group(1) == 'UtcTime':
            sheet['D'+str(flag)] = match.group(2)
            flag = flag + 1
        #print(flag)
```
